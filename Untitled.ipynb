```python
# Title
## Step 1: Importing Required Libraries
import pandas as pd
import numpy as np
import os
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import joblib
from scipy.sparse import hstack, csr_matrix
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from itertools import cycle
from sklearn.neighbors import KNeighborsClassifier

from sklearn.naive_bayes import GaussianNB
```

```python
# Create directory to save models
MODEL_DIR = "models"
Path(MODEL_DIR).mkdir(exist_ok=True)
```

```python
def load_dataset(file_path):
    """Load the dataset from a CSV file."""
    try:
        df = pd.read_csv(file_path)
        print(f"Dataset loaded successfully with shape: {df.shape}")
        return df
    except FileNotFoundError:
        print(f"Error: File {file_path} not found.")
        return None
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return None
```

```python
def preprocess_data(df, is_train=True):
    """Preprocess the dataset for training or inference."""
    import os
    import joblib
    import numpy as np
    from sklearn.preprocessing import LabelEncoder, StandardScaler
    from scipy.sparse import hstack, csr_matrix

    df = df.copy()

    # Drop or fill missing values
    if is_train:
        df = df.dropna()
    else:
        df.fillna(df.mean(numeric_only=True), inplace=True)

    # Drop ID column if it exists
    if 'ID' in df.columns:
        df = df.drop('ID', axis=1)

    # Define base categorical columns
    categorical_cols = [
        'Listing Confirmation Method', 'Direction', 'Parking Availability',
        'Real Estate Agency', 'Brokerage Platform', 'Listing Weekday'
    ]
    if is_train and 'False Listing' in df.columns:
        categorical_cols.append('False Listing')

    if is_train:
        le_dict = {}
        for col in categorical_cols:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col].astype(str))
            le_dict[col] = le
        joblib.dump(le_dict, os.path.join(MODEL_DIR, 'label_encoders1.pkl'))
    else:
        le_dict = joblib.load(os.path.join(MODEL_DIR, 'label_encoders1.pkl'))
        for col in categorical_cols:
            le = le_dict[col]

            # Handle unseen labels
            known_classes = set(le.classes_)
            df[col] = df[col].astype(str).apply(lambda x: x if x in known_classes else 'Unknown')

            # Ensure 'Unknown' is in the encoder
            if 'Unknown' not in le.classes_:
                le.classes_ = np.append(le.classes_, 'Unknown')

            df[col] = le.transform(df[col])

    # Target encoding (only in training)
    if is_train:
        le_target = LabelEncoder()
        df['False Listing'] = le_target.fit_transform(df['False Listing'].astype(str))
        joblib.dump(le_target, os.path.join(MODEL_DIR, 'target_label_encoder1.pkl'))
        y = df['False Listing']
    else:
        y = None

    # Numerical columns to scale
    numerical_cols = [
        'Deposit Amount', 'Monthly Rent', 'Exclusive Area', 'Unit Floor',
        'Total Floors', 'Number of Rooms', 'Number of Bathrooms',
        'Total Parking Spaces', 'Maintenance Fee', 'Listing Year',
        'Listing Month', 'Listing Day', 'Days Since Listed'
    ]

    # Scaling
    if is_train:
        scaler = StandardScaler()
        numerical_features = scaler.fit_transform(df[numerical_cols])
        joblib.dump(scaler, os.path.join(MODEL_DIR, 'scaler1.pkl'))
    else:
        scaler = joblib.load(os.path.join(MODEL_DIR, 'scaler1.pkl'))
        numerical_features = scaler.transform(df[numerical_cols])

    # Combine categorical + numerical
    other_features = df[categorical_cols].values
    X = hstack([csr_matrix(other_features), csr_matrix(numerical_features)])

    return df, X, y
```

```python
def perform_eda(df):
    """Perform exploratory data analysis and generate plots."""
    plt.figure(figsize=(15, 10))
    
    # Plot 1: Price distribution
    plt.subplot(2, 3, 1)
    sns.distplot(df['Monthly Rent'], kde=True)
    plt.title('Distribution of Monthly Rent')

    # Plot 2: Ratings vs Price
    plt.subplot(2, 3, 2)
    sns.scatterplot(x='Total Floors', y='Number of Rooms', data=df)
    plt.title('Number of Rooms vs Total Floors')

    # Plot 3: Category vs Price
    plt.subplot(2, 3, 3)
    sns.barplot(x='False Listing', y='Monthly Rent', data=df)
    plt.title('Monthly Rent by False Listing')
    plt.xticks(rotation=45)

    # Plot 4: Price by Top Cities
    top_cities = df['Parking Availability'].value_counts().index[:5]
    plt.subplot(2, 3, 4)
    sns.boxplot(x='Parking Availability', y='Exclusive Area', data=df[df['Parking Availability'].isin(top_cities)])
    plt.title('Exclusive Area VS Parking Availability')
    plt.xticks(rotation=45)

    # Plot 5: Correlation heatmap
    plt.subplot(2, 3, 5)
    corr = df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    
    plt.tight_layout()
    plt.savefig('eda_plots.png')
    plt.show()
    print("EDA plots saved as 'eda_plots.png'")
```

```python
def train_test_split_data(X, y, test_size=0.2, random_state=42):
    """Split the data into training and testing sets."""
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )
    print(f"Training shape: {X_train.shape}, Testing shape: {X_test.shape}")
    return X_train, X_test, y_train, y_test
```

```python
# Global metric tracking
metrics_df = pd.DataFrame(columns=['Algorithm', 'Accuracy', 'Precision', 'Recall', 'F1-Score'])

def Calculate_Metrics(algorithm, y_pred, y_test, y_score=None):
    le_target = joblib.load(os.path.join(MODEL_DIR, 'target_label_encoder1.pkl'))
    categories = le_target.classes_

    # Metrics
    acc = accuracy_score(y_test, y_pred) * 100
    prec = precision_score(y_test, y_pred, average='macro') * 100
    rec = recall_score(y_test, y_pred, average='macro') * 100
    f1 = f1_score(y_test, y_pred, average='macro') * 100

    metrics_df.loc[len(metrics_df)] = [algorithm, acc, prec, rec, f1]

    print(f"{algorithm} Accuracy: {acc:.2f}%")
    print(f"{algorithm} Precision: {prec:.2f}%")
    print(f"{algorithm} Recall: {rec:.2f}%")
    print(f"{algorithm} F1-Score: {f1:.2f}%")
   # print(f"\nClassification Report:\n{classification_report(y_test, y_pred, target_names=categories)}")
    print(f"\nClassification Report:\n{classification_report(y_test, y_pred, labels=range(len(categories)), target_names=categories)}")
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, xticklabels=categories, yticklabels=categories, cmap="Blues", fmt="g")
    plt.title(f'{algorithm} - Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig(f"results/{algorithm.replace(' ', '_')}_confusion_matrix.png")
    plt.show()
    
    # ROC Curve Plot (if y_score is provided)
    if y_score is not None:
        y_test_bin = label_binarize(y_test, classes=range(len(categories)))

        plt.figure(figsize=(10, 8))

        # Handle Binary Classification Case
        if len(categories) == 2:
            # Use probability of positive class (1)
            fpr, tpr, _ = roc_curve(y_test, y_score[:, 1])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'{categories[1]} vs {categories[0]} (AUC = {roc_auc:.2f})')
        
        # Handle Multi-Class Case
        else:
            fpr = dict()
            tpr = dict()
            roc_auc = dict()
            for i in range(len(categories)):
                fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
                roc_auc[i] = auc(fpr[i], tpr[i])
                plt.plot(fpr[i], tpr[i], label=f'Class {categories[i]} (AUC = {roc_auc[i]:.2f})')

        plt.plot([0, 1], [0, 1], 'k--', label='Random')
        plt.title(f"{algorithm} ROC Curve")
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend(loc='lower right')
        plt.grid(True)
        plt.tight_layout()
        plt.savefig(f"results/{algorithm.replace(' ', '_')}_roc_curve.png")
        plt.show()

```

```python
def train_logistic_regression(X_train, y_train, X_test, y_test):
    model_path = os.path.join(MODEL_DIR, 'logistic_regression1.pkl')
    
    if os.path.exists(model_path):
        print("Loading Logistic Regression model...")
        model = joblib.load(model_path)
    else:
        print("Training Logistic Regression model...")
        model = LogisticRegression()
        model.fit(X_train, y_train)
        joblib.dump(model, model_path)
        print(f"Model saved to {model_path}")
    
    y_pred = model.predict(X_test)
    y_score = model.predict_proba(X_test)
    Calculate_Metrics("Logistic Regression", y_pred, y_test, y_score)
    return model



def train_knn_classifier(X_train, y_train, X_test, y_test):
    model_path = os.path.join(MODEL_DIR, 'knn_classifier1.pkl')
    
    if os.path.exists(model_path):
        print("Loading KNN model...")
        model = joblib.load(model_path)
    else:
        
        print("Training KNN model...")
        model = KNeighborsClassifier(n_neighbors=10)  # You can tune k as needed
        model.fit(X_train, y_train)
        joblib.dump(model, model_path)
        print(f"Model saved to {model_path}")
    
    y_pred = model.predict(X_test)
    y_score = model.predict_proba(X_test)
    Calculate_Metrics("KNN model", y_pred, y_test, y_score)
    return model

def train_hybrid_extra_tree_ann(X_train, y_train, X_test, y_test, top_k_features=100):
    model_path = os.path.join(MODEL_DIR, 'hybrid_extra_tree_ann1.pkl')
    et_path = os.path.join(MODEL_DIR, 'feature_selector_extra_tree1.pkl')

    if os.path.exists(model_path):
        print("Loading Hybrid ExtraTree + ANN model...")
        model, indices = joblib.load(model_path)
        X_test_reduced = X_test[:, indices]
    else:
        print("Training Extra Trees for feature selection...")
        et_model = ExtraTreesClassifier(n_estimators=100, random_state=42)
        et_model.fit(X_train, y_train)
        joblib.dump(et_model, et_path)

        importances = et_model.feature_importances_
        indices = np.argsort(importances)[::-1][:top_k_features]
        print(f"Selected top {top_k_features} features from Extra Trees.")

        X_train_reduced = X_train[:, indices]
        X_test_reduced = X_test[:, indices]

        print("Training ANN (MLPClassifier)...")
        model = MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=300, random_state=42)
        model.fit(X_train_reduced, y_train)

        joblib.dump((model, indices), model_path)
        print(f"Hybrid model saved to {model_path}")

    y_pred = model.predict(X_test_reduced)
    y_score = model.predict_proba(X_test_reduced)
    Calculate_Metrics("Hybrid AI (ANN and ET) Model", y_pred, y_test, y_score)

```

```python
def predict_class_only_real_estate(sample_df, model_name='logistic_regression1.pkl'):
    import os
    import joblib
    import scipy.sparse
    from scipy.sparse import hstack, csr_matrix

    model_path = os.path.join("models", model_name)
    loaded = joblib.load(model_path)

    _, X_sample, _ = preprocess_data(sample_df.copy(), is_train=False)

    if model_name == 'hybrid_extra_tree_ann1.pkl':
        estimator, selected_indices = loaded

        if isinstance(X_sample, scipy.sparse.spmatrix):
            X_sample = X_sample.tocsr()

        # ✅ Fix: Pad if not enough features, and convert to CSR
        if X_sample.shape[1] <= max(selected_indices):
            missing = max(selected_indices) + 1 - X_sample.shape[1]
            X_sample = hstack([X_sample, csr_matrix((X_sample.shape[0], missing))]).tocsr()

        # Now safely slice
        X_sample = X_sample[:, selected_indices]
        y_pred = estimator.predict(X_sample)
    else:
        model = loaded
        y_pred = model.predict(X_sample)

    return y_pred
```
